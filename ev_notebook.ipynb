```python
import pandas as pd
import re
```

```python
ev_data = "./Resources/Electric_Vehicle_Population_Data.csv"
df = pd.read_csv(ev_data)
df.head()
```

```python
df1 = df.drop(['Postal Code', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility','Base MSRP', 'Legislative District', 'DOL Vehicle ID', 'Electric Utility', '2020 Census Tract'], axis = 1)
df1.head()
```

```python
df1.info()
```

```python
clean_df = df1.dropna()
clean_df.head()
```

```python
clean_df = clean_df.replace(to_replace='\(', value ="", regex = True)
clean_df
clean_df = clean_df.replace(to_replace='\)', value ="", regex = True)
clean_df
```

```python
clean_df.info()
```

```python
clean_df = clean_df.rename(columns = {'Vehicle Location' : "location"}) 
```

```python
clean_df[['point', 'Latitude', 'Longitude']] = clean_df.location.str.split(" ", expand = True)
clean_df.head()
```

```python
clean_df.info()
```

```python
clean_df['Latitude']= clean_df['Latitude'].astype('float')
clean_df
clean_df['Longitude']= clean_df['Longitude'].astype('float')
clean_df
```

```python
clean_df.info()
```

```python
clean_df = clean_df.drop(['location', 'point'], axis = 1)
```

```python
clean_df
```

```python
clean_df = clean_df.rename(columns = {'VIN (1-10)' : "vin", "County": "county", "City": "city", "State": "state", "Model Year": "model_year", "Make" : "make", "Model": "model", "Electric Vehicle Type" : "vehicle_type" , "Electric Range" : "range", "Latitude" : "latitude", "Longitude" : "longitude"})
clean_df
```

```python
clean_df.to_csv("output/clean_df.csv", encoding='utf8', index=False)
```
